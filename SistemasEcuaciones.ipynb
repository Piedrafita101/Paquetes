**Sistema de Ecuaciones lineales**

```python
import numpy as np
```

```python
import torch
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000) # inicio, final, n punto
```

```python
y1 = 3*x
```

```python
y2 = 1 + (5*x)/2
```

```python
from matplotlib.lines import lineStyles
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0, 3])
ax.set_ylim([0, 8])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c = 'red')
plt.axvline(x=2, color='purple', linestyle='--')
_ = plt.axhline(y=6,color='purple', linestyle = '--')
```

**Otro sistema**

```python
y1= -5 + ((2*x)/3)
```

```python
y2 = (7-2*x)/5
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')

# Añadimos los ejes x e y:
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

ax.set_xlim([-2, 10])
ax.set_ylim([-6, 4])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c= 'red')
plt.axvline(x=6, color='purple', linestyle='--')
_ =plt.axhline(y=-1, color='purple', linestyle='-.')
```

##LA NORMA DE FROEBENIUS

```python
X = np.array([[1,2], [3, 4]])
X
```

```python
(1**2 + 2**2 + 3**2 + 4**2)**(1/2)
```

```python
np.linalg.norm(X) # La misma función que para la norma vectotial L2
```

```python
X_pt = torch.tensor([[1,2], [3, 4.]]) # torch.norm() solamente soporta floats
```

```python
torch.norm(X_pt)
```

```python
import tensorflow as tf
X_tf = tf.Variable([[1.,2.], [3., 4.]]) # tf.norm() solamente soporta floats
```

```python
tf.norm(X_tf)
```

```python
A = np.array([[3,4], [5, 6], [7,8]])
A
```

```python
B = np.array([[1,9], [2, 0]])
B
```

```python
np.dot(A, B)
```

**La multiplicación de matrices No es Conmuntativa**

```python
B_pt = torch.from_numpy(B)
B_pt
```

```python
A_pt = torch.from_numpy(A)
A_pt
```

```python
torch.matmul(A_pt, B_pt) # no se necesita cambiar la función, a diferencia de TF
```

```python
B_pt = torch.tensor([[1, 2], [9, 0]]).T
B_pt
```

```python
torch.matmul(A_pt, B_pt)
```

```python
B_tf = tf.convert_to_tensor(B, dtype=tf.float32)
B_tf
```

```python
A_tf = tf.convert_to_tensor(A, dtype=tf.float32)
A_tf
```

```python
tf.matmul(A_tf, B_tf)
```
